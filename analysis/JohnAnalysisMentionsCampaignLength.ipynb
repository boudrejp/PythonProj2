```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

# question 2: Do song mentions tend to correlate to how long a candidate was in the race?
# use secondary data sources to approximate how long each candidate was in race
# this will involve converting the dates into datetime objects for addition, subtraction
# then create some type of list of number of unique entries by candidate
# side by side bar chart with multiple axes


campaign_len = pd.read_csv("../data/campaign_length.csv")
lyric_data = pd.read_csv("../data/HipHopLyrics.csv")
```

```python
campaign_len
```

```python
lyric_data.head()
```

```python
wanted_frame = lyric_data[['candidate', 'song']].groupby('candidate').nunique()
wanted_frame['name'] = wanted_frame.index
wanted_frame
```

```python
# preprocess for a ghetto join
campaign_len.drop('Unnamed: 0', axis = 1, inplace = True)

campaign_len.rename(index = str, columns={'Trump': 'Donald Trump', 'Cruz':'Ted Cruz', 'Kasich':'John Kasich', \
                     'Clinton':'Hillary Clinton', 'Sanders':'Bernie Sanders'}, inplace = True)
camp_len = campaign_len.transpose()
camp_len['name'] = camp_len.index
```

```python
merged = pd.merge(wanted_frame, camp_len, left_on='name', right_on='name')
merged.rename(index = str, columns={'song':'mentions', '0':'campaign_length'}, inplace = True)
merged.drop('candidate', axis = 1, inplace = True)
merged.index = merged.name
merged.drop('name', axis = 1, inplace = True)
new_merged = pd.concat([merged,pd.DataFrame({'mentions': [0], 'campaign_length':[357]})])

as_list = new_merged.index.tolist()
idx = as_list.index(0)
as_list[idx] = 'John Kasich'
new_merged.index = as_list


new_merged
```

```python
plt.scatter(x = new_merged.campaign_length, y = new_merged.mentions)
plt.xlabel("Campaign length (days)")
plt.ylabel("Song mentions")
plt.title("Song mentions with respect to campaign length")
plt.show()
```
