```python
import csv
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

with open('HipHopLyrics.csv','rt',encoding='utf-8') as CSVread:
    has_header = csv.Sniffer().has_header(CSVread.read(1024))
    CSVread.seek(0)
    CSV_read = csv.DictReader(CSVread,fieldnames=['id','candidate','song'
                                                  ,'artist','sentiment',
                                                  'theme','album_release_date',
                                                  'line','url'])
    if has_header:
        next(CSV_read)
    list_data = [dict(row) for row in CSV_read]

dict_stats = {}
for j in range(1989,2017):
    count_pos = 0
    count_neg = 0
    count_neu = 0
    list_temp = [i['sentiment'] for i in list_data if i['candidate'] == 'Donald Trump' and i['album_release_date'] == str(j)]
    #print(list_temp)
    count_pos = list_temp.count('positive')
    count_neg = list_temp.count('negative')
    count_neu = list_temp.count('neutral')
    dict_stats.update({j:(count_pos,count_neg,count_neu)})
    
print(dict_stats)
print("Trump mentiones histogram")
pos_hist = pd.Series([dict_stats.get(i)[0] for i in dict_stats], index=dict_stats.keys())
# print(pos_hist)
ax1 = pos_hist.plot(label='positive')

neg_hist = pd.Series([dict_stats.get(i)[1] for i in dict_stats], index=dict_stats.keys())
ax2 = neg_hist.plot(label='negative')

neu_hist = pd.Series([dict_stats.get(i)[2] for i in dict_stats], index=dict_stats.keys())
ax3 = neu_hist.plot(label='neutral')

plt.legend()
```
